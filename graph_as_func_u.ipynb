```python
######## If you have an .npz file and a .npy file use this code ############



import numpy as np
import matplotlib.pyplot as plt

# Load data
predictions = np.load(r'/home/fabio/work/HM_and_AI_models/VAE_Model/predictions.npz')
original_data = np.load(r"/home/fabio/work/HM_and_AI_models/VAE_Model/data/hm_data_10.npy")

# Define time limits
left_limit, right_limit = 10, 11

# Extract predicted and original data
u_profile_pred = predictions["predictions"][left_limit:right_limit]
u_profile_original = original_data[left_limit:right_limit, 1, 61]

# Define time steps
time_steps = np.arange(left_limit, right_limit)

# Plot time series
plt.figure(figsize=(10, 5))
plt.plot(time_steps, u_profile_pred, label="Predicted", linestyle="--")
plt.plot(time_steps, u_profile_original, label="Original", alpha=0.7)
plt.xlabel("Time step")
plt.ylabel("U (m/s)")
plt.title("Time Series of U at Vertical Level 61")
plt.legend()
plt.show()

# Select time steps for successive comparison
time_step, next_time_step = 20, 21

# Define vertical levels
z_levels_U = np.linspace(0, 70, 23)
z_levels_RePsi = np.linspace(0, 70, 24)
z_levels_ImPsi = np.linspace(0, 70, 25)

# Extract data for U, Re(Psi), and Im(Psi)
U_current_pred = np.full_like(z_levels_U, predictions["predictions"][time_step])
U_next_pred = np.full_like(z_levels_U, predictions["predictions"][next_time_step])
U_current_orig, U_next_orig = original_data[time_step, 1, 51:74], original_data[next_time_step, 1, 51:74]
RePsi_current_pred, RePsi_next_pred = original_data[time_step, 1, 0:24], original_data[next_time_step, 1, 0:24]
RePsi_current_orig, RePsi_next_orig = original_data[time_step, 1, 0:24], original_data[next_time_step, 1, 0:24]
ImPsi_current_pred, ImPsi_next_pred = original_data[time_step, 1, 25:50], original_data[next_time_step, 1, 25:50]
ImPsi_current_orig, ImPsi_next_orig = original_data[time_step, 1, 25:50], original_data[next_time_step, 1, 25:50]

# Compute differences
U_diff_pred, U_diff_orig = U_next_pred - U_current_pred, U_next_orig - U_current_orig
RePsi_diff_pred, RePsi_diff_orig = RePsi_next_pred - RePsi_current_pred, RePsi_next_orig - RePsi_current_orig
ImPsi_diff_pred, ImPsi_diff_orig = ImPsi_next_pred - ImPsi_current_pred, ImPsi_next_orig - ImPsi_current_orig

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 15), sharey='row')

def plot_variable(ax, pred_current, orig_current, pred_next, orig_next, levels, xlabel, title):
    ax.plot(pred_current, levels, 'o--', label="Predicted Current")
    ax.plot(orig_current, levels, 'x', label="Original Current")
    ax.plot(pred_next, levels, 's--', label="Predicted Next")
    ax.plot(orig_next, levels, 'd', label="Original Next")
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel("Vertical Levels (km)")
    ax.legend()

def plot_difference(ax, diff_pred, diff_orig, levels, xlabel, title):
    ax.plot(diff_pred, levels, 'o--r', label="Predicted")
    ax.plot(diff_orig, levels, 'xb', label="Original")
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.legend()

# Plot U, Re(Psi), Im(Psi)
plot_variable(axes[0, 0], U_current_pred, U_current_orig, U_next_pred, U_next_orig, z_levels_U, "U (m/s)", "U vs Height (Current & Next)")
plot_difference(axes[0, 1], U_diff_pred, U_diff_orig, z_levels_U, "U (m/s)", "U Difference (Next - Current)")
plot_variable(axes[1, 0], RePsi_current_pred, RePsi_current_orig, RePsi_next_pred, RePsi_next_orig, z_levels_RePsi, "Re(Psi)", "Re(Psi) vs Height (Current & Next)")
plot_difference(axes[1, 1], RePsi_diff_pred, RePsi_diff_orig, z_levels_RePsi, "Re(Psi)", "Re(Psi) Difference (Next - Current)")
plot_variable(axes[2, 0], ImPsi_current_pred, ImPsi_current_orig, ImPsi_next_pred, ImPsi_next_orig, z_levels_ImPsi, "Im(Psi)", "Im(Psi) vs Height (Current & Next)")
plot_difference(axes[2, 1], ImPsi_diff_pred, ImPsi_diff_orig, z_levels_ImPsi, "Im(Psi)", "Im(Psi) Difference (Next - Current)")

# Adjust layout and save
plt.tight_layout()
plt.savefig("U_RePsi_ImPsi_Successive_TimeSteps_Comparison.png")
plt.show()

# Debugging prints
print(predictions["predictions"].shape) 
print(original_data.shape)

```
