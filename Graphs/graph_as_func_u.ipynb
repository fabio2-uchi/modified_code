```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data; shape = (300000, 2, 75)
data = np.load(r"/home/fabio/work/HM_and_AI_models/VAE_Model/x_stoch.npy")

in_strong = 5000
in_weak = 21000
in_transition = 14500

#U speed graph
u_profile = data[in_strong, 0, 51:74]
z_levels = np.arange(51, 74)
fig, ax = plt.subplots()
ax.plot(u_profile, z_levels, marker='o')
ax.set_xlabel("U (m/s)")
ax.set_ylabel("Vertical Levels")
ax.set_title("Profile of U vs Height (Last 25 Levels)")
plt.savefig("U_strong_state")
plt.show()


## IM Psi Graph
u_profile = data[in_strong, 0, 25:50]
z_levels = np.arange(25, 50)
fig, ax = plt.subplots()
ax.plot(u_profile, z_levels, marker='o')
ax.set_xlabel("IM Psi")
ax.set_ylabel("Vertical Levels")
ax.set_title("Profile of IM Psi vs Height (Middle 25 Levels)")
plt.savefig("IM_psi_strong_state")
plt.show()


## Re Psi Graph
u_profile = data[in_strong, 0, 0:24]
z_levels = np.arange(0, 24)
fig, ax = plt.subplots()
ax.plot(u_profile, z_levels, marker='o')
ax.set_xlabel("RE Psi")
ax.set_ylabel("Vertical Levels")
ax.set_title("Profile of Re Psi vs Height (First 25 Levels)")
plt.savefig("Re_psi_strong_state")
plt.show()


```
